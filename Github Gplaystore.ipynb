### Changing DataType of coln 

```python
df['Reviews']=df['Reviews'].astype(float)
```

```python
df.info()
```

```python
df['Size'].value_counts()
```

```python
def convert_size(x):
    if 'k' in x:
        x=x.replace('k','')
        return float(x)/1024
    elif 'M' in x:
        x=x.replace('M','')
        return float(x)
    elif x=='Varies with device':
        return np.nan
    else:
        return float(x)
```

```python
df['Size']=df['Size'].map(convert_size)
```

```python
df.info()
```

### Working on Price coln -- DataType / removing $ from Price

```python
df['Price']=df['Price'].str.replace("$"," ")
```

```python
df['Price'].value_counts() # To see if $ is replaced to Empty
```

```python
df['Price']=df['Price'].astype(float) #To change the datatype
```

### Replacing values and DataTpe in 'Installs' col
